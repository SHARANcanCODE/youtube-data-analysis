```python
import kagglehub
import pandas as pd

path = kagglehub.dataset_download("surajjha101/top-youtube-channels-data")
print("Path to dataset files:", path)

csv_file_path = path + "/most_subscribed_youtube_channels.csv"
df = pd.read_csv(csv_file_path)

display(df.head())

columns_to_clean = ['subscribers', 'video views']

for col in columns_to_clean:
    df[col] = df[col].astype(str).str.replace(',', '', regex=False).astype(float)

display(df.head())
```

```python
import plotly.express as px

fig = px.pie(df, values='subscribers', names='category', title='YouTube Categories')
fig.show()
```

```python
import plotly.express as px

fig = px.box(df, y='started', title='Years Started')
fig.show()
```

```python
import plotly.express as px
fig = px.histogram(df, x='subscribers', title='Subscriber Count')
fig.show()
```
